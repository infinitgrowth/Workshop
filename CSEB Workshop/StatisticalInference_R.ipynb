```r
library(zoo)
library(dplyr)
#Would love to convert everything to csv but sometimes it's just in xlsx
heart = read.csv("heart.csv")
```

```r
#t-test
t.test(na.aggregate(heart[heart$Sex =="Male","Weight"]), 
       y = na.aggregate(heart[heart$Sex == "Female","Weight"]), 
       var.equal=TRUE)
```

```r
#Anova
oneway.test(na.aggregate(heart$Weight)~Sex, data = heart, na.action=na.omit, var.equal=FALSE)
```
