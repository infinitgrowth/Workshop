```r
library(e1071)
heart = read.csv("heart.csv")
```

```r
summary(heart$Systolic)
sd(heart$Systolic) #standard deviation
var(heart$Systolic) #variance
range(heart$Systolic)
IQR(heart$Systolic) #interquartile range
skewness(heart$Systolic)
kurtosis(heart$Systolic)
sum(is.na(heart$Systolic)) #missing values
```

```r
hist(heart$Systolic, col = rainbow(12), main = "Systolic") # col = rainbow(12) for a little color
#split histogram by categorical variable
par(mfrow = c(1,2))
hist(heart$Systolic[heart$Status=="Alive"], main = "Systolic by Status", col = rainbow(12))
hist(heart$Systolic[heart$Status=="Dead"], main = "", col = rainbow(12))
```

```r
boxplot(heart$Systolic)
boxplot(Systolic ~ Status, data=heart) #split by classification variable
```

```r
qqnorm(heart$Systolic, pch=18, main="Systolic")
qqline(heart$Systolic, col = "red", lwd=3)
```

```r
barplot(table(heart$Status)) # plot of categorical frequency
```
