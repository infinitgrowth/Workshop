```python
import pandas as pd
import scipy as sp
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as seaborn
```

```python
heart = pd.read_csv('heart.csv')
```

```python
#main descriptive statistics
sp.stats.describe(heart.Systolic)
```

```python
import seaborn;seaborn.set()
plt.hist(heart.Systolic, edgecolor='black')
plt.title("Systolic Frequency")
plt.xlabel("Systolic")
plt.ylabel("Frequency")
plt.show();
```

```python
# Setting Figure
fig=plt.figure()
ax = fig.add_subplot(1,1,1)
# Triggering Boxplot Chart
ax.boxplot(heart['Systolic'],showfliers=True, flierprops = dict(marker='o', markerfacecolor='green', markersize=3,linestyle='none'))
# Setting Plot Title
plt.title('Systolic Boxplot')
# Show Plot
plt.show()
```

```python
# Choose only publication and Age variable
heartbp = heart.ix[:,['Systolic','Status']]
# Boxplot
heartbp.boxplot(by='Status',rot = 90,showfliers=True, flierprops = dict(marker='o', markerfacecolor='green', markersize=3,linestyle='none'))
plt.title('Systolic Boxplot by Status')
# Show Plot
plt.show()
```

```python
import pylab
stats.probplot(heart['Systolic'], dist="norm", plot=pylab)
pylab.show()
```
